```python
discrete_columns = ['Gender', 'Geography', 'Tenure', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'Exited', 'Complain', 'Satisfaction Score', 'Card Type']
```

```python
num_cols = len(discrete_columns)
# Calculate the optimal number of rows and columns for the subplots
num_rows = (num_cols + 1) // 2  # Ensure enough rows to fit all columns
num_cols_grid = min(num_cols, 2) # Display at most 2 columns of subplots

# Adjust the figure size for better readability
plt.figure(figsize=(12, 5 * num_rows)) # Adjust width and height as needed
palettes = ['Blues_r', 'Greens_r', 'Purples_r', 'Oranges_r', 'Reds_r',
            'BuPu_r', 'GnBu_r', 'PuRd_r', 'YlGnBu_r', 'RdPu_r']

for i, col in enumerate(discrete_columns):
    plt.subplot(num_rows, num_cols_grid, i + 1)
    sns.countplot(x=col, data=Data, hue=col, palette=palettes[i % len(palettes)], legend=False)
    plt.title(f'Distribution of {col}') # Add a title to each subplot
    plt.xlabel(col) # Label the x-axis
    plt.ylabel('Count') # Label the y-axis
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability if they are long

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```
